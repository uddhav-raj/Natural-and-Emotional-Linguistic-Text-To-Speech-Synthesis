```python
import librosa
import matplotlib.pyplot as plt
import librosa.display
from IPython.display import Audio
```

```python
import numpy as np
import math
```

```python
angry_path = "./data/angry/03-01-05-01-01-01-01.wav"
happy_path = "./data/happy/03-01-03-01-01-01-01.wav"
sad_path = "./data/sad/03-01-04-01-01-01-01.wav"
fearful_path = "./data/fearful/03-01-06-01-01-01-01.wav"
neutral_path = "./data/neutral/03-01-01-01-01-01-01.wav"
disgust_path = "./data/disgust/03-01-07-01-01-01-01.wav"
surprised_path = "./data/surprised/03-01-08-01-01-01-01.wav"
calm_path = "./data/calm/03-01-02-01-01-01-01.wav"
```

```python
y, sr = librosa.load(angry_path, duration=10)
Audio(data=y, rate=sr)
```

```python
S, phase = librosa.magphase(librosa.stft(y))
rms = librosa.feature.rmse(S=S)
plt.figure()
plt.subplot(2, 1, 1)
plt.semilogy(rms.T, label='RMS Energy_')
plt.xticks([])
plt.xlim([0, rms.shape[-1]])
plt.legend(loc='best')
plt.title('Spectogram_Angry')
plt.subplot(2, 1, 2)
librosa.display.specshow(librosa.amplitude_to_db(S, ref=np.max), y_axis='log', x_axis='time')
plt.title('log Power spectrogram')
plt.show()
```

```python
rms = librosa.feature.rmse(y=y)
rmse_angry=0
for rm in rms:
    for num in rm:
        rmse_angry = rmse_angry + num*num
    break
rmse_angry = rmse_angry/rms.size
rmse_angry = math.sqrt(rmse_angry)
rmse_angry
```
