```python
import librosa
```

```python
angry_path = "./data/angry/03-01-05-01-01-01-01.wav"
happy_path = "./data/happy/03-01-03-01-01-01-01.wav"
sad_path = "./data/sad/03-01-04-01-01-01-01.wav"
fearful_path = "./data/fearful/03-01-06-01-01-01-01.wav"
neutral_path = "./data/neutral/03-01-01-01-01-01-01.wav"
disgust_path = "./data/disgust/03-01-07-01-01-01-01.wav"
surprised_path = "./data/surprised/03-01-08-01-01-01-01.wav"
calm_path = "./data/calm/03-01-02-01-01-01-01.wav"
```

```python
import matplotlib.pyplot as plt
```

```python
import librosa.display
```

```python
from IPython.display import Audio
```

```python
y, sr = librosa.load(angry_path, duration=10)
Audio(data=y, rate=sr)
```

```python
y, sr = librosa.load(angry_path, duration=20)
plt.figure(figsize=(15, 8))
plt.subplot(1, 2, 1)
librosa.display.waveplot(y, sr=sr)
plt.title('Angry_Waveform')
plt.subplot(1, 2, 2)
C = librosa.feature.chroma_cqt(y=y, sr=sr)
librosa.display.specshow(C, y_axis='chroma')
plt.colorbar()
plt.title('Angry_Chromagram')
plt.show()
```
